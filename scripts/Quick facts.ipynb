```python
proot = "C:/Users/rwilliams/Documents/portal/"
xfile = proot + "content/quick-facts.xls"
xfdata = pd.ExcelFile(xfile)
xfdata.parse("Sheet1", skiprows=1, index="Field")
```

```python
def diceData(xlfile,sheet,country):
    data = xlfile.parse(sheet, skiprows=1)
    data = data[["Field",country]]
    POP = 0 # This is fragile code that relies on the rows being in a certain order
    LA = 2  # I couldn't convince pandas to do it a better way
    density = '%.2f' % (data.loc[POP][country]/data.loc[LA][country])
    s = pd.Series(["Density",density], index=['Field',country])
    #df = pd.DataFrame([["Name","Density"][country,density]],index=[8])
    data = data.append(s, ignore_index=True)
    return data
country = "Anguilla"
data = diceData(xfdata,"Sheet1",country)
data
#data.index
#data = data.stack().unstack(level=0)


#density = data["Population"] / data ["Land Area"]
```
