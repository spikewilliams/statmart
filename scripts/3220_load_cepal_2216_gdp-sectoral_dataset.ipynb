This script imports CEPALStat data for indicator \#2216, GDP by sector.

This script will import data from all files listed in "data/facts/gen2/cepalstat/_gdp-sec.csv".

```python
import os.path
import pandas as pd
from hashlib import sha1
import pymysql
import pandas.io.sql as psql

import statmart_util as smu

from statmart_config import *

gen2Dir = statmart_facts_gen2 + "cepalstat/gdp-sec/"

prefix = "gdp-sec"
suffix = "cepalstat"
indicator = "2216"
```

Bring in all the data and metadata using the statmart_util.get_gen2_data function. The data is returned as an array of tupples. Each tupple contains the "series key", the ISO3 value of the country, the meta DataFrame, and the dataset DataFrame. 

```python
data_list = smu.get_gen2_data(gen2Dir, prefix)
data_list[0:5] # have a look at a sample our data
```

Insert the data from data_list into the MySQL database

```python
with smu.get_db_connection() as cursor:
    for (series_key, iso3, meta, data) in data_list:
        cursor.execute("SELECT ID FROM location WHERE iso3 = %s AND divisionname IS NULL AND city IS NULL", [iso3])
        locationid = list(cursor)[0][0]
        
        series_fields = "identifier, dataset, file, filehash, category, type, name, description"
        insert_series = "INSERT INTO SERIES (" + series_fields +") VALUES (%s, %s, %s, %s, %s, %s, %s, %s)" 
        values = [series_key]
        for field in series_fields.split(",")[1:]:
            values.append(meta[field.strip()])
        cursor.execute(insert_series, values)
        
        for date in data.index:
            cursor.execute("SELECT ID FROM date WHERE year = %s AND month IS NULL AND day IS NULL", [str(date)])
            dateid = list(cursor)[0][0]
            observation_fields = "series, locationid, dateid, value" 
            #full list of fields: `OBSERVATIONID`, `SERIESID`, `LOCATIONID`, `DATEID`, `NOTEID`, `DESCRIPTIONID`, `VALUE`, `UNITID`, `STATUSID`
            insert_observation = "INSERT INTO observation(" + observation_fields + ") VALUES (%s, %s, %s, %s)"
            values  = [series_key, locationid, dateid, str(data.ix[date]["value"])]
            cursor.execute(insert_observation, values)
        cursor.execute("COMMIT")
```

Here is an example of how to make an SQL query to retrieve data in the series:

```python
df = ""
with smu.get_db_connection() as cursor:
    query = """
            SELECT L.iso3, L.countryname, D.year, O.value 
            FROM observation as O 
            INNER JOIN date as D    ON O.dateid = D.id
            INNER JOIN location as L    ON O.locationid = L.id
            WHERE O.series like 'gdp-sec-cstrn\_%\_cepalstat'
            ORDER BY D.year
            """
    df = psql.frame_query(query, con=cursor.connection)
    
df
```
