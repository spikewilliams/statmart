This script imports DESA population data into the database. The data can be found in the /facts/gen2/desa directory. The two data files per country are pop\__iso3_\_desa.csv, containing the time series, and pop\__iso3_\_desa_meta.csv, containing meta data.

This script only imports DESA data for Caribbean countries, as listed in country_caribbean.csv.

```python
import os.path
import pandas as pd
from hashlib import sha1
import pymysql
import pandas.io.sql as psql

import statmart_util as smu

from statmart_config import *

gen2Dir = statmart_facts_gen2 + "desa/"

prefix = "pop"
suffix = "desa"
```

Load up the list of Caribbean countries...

```python
carib_country_path = statmart_dimensions_gen2 + "cepalstat/country_caribbean.csv"
countries = pd.read_csv(carib_country_path, encoding="utf-8", usecols=["iso3"], squeeze=True)
```

Load the data for each country in the list... 

```python
data_list = [] 
for c in countries:
    metafile = gen2Dir + prefix + "_" + c.lower() + "_" + suffix + "_meta.csv"
    if os.path.isfile(metafile):
        meta = pd.read_csv(metafile, encoding="utf-8", index_col=["key"], squeeze=True)
        datafile = gen2Dir + prefix + "_" + c.lower() + "_" + suffix + ".csv"
        data = pd.read_csv(datafile, encoding="utf-8", index_col=["year"], squeeze=True)
        data_list.append((c,meta,data))
```

Load up the list of Caribbean countries...

```python
with smu.get_db_connection() as cursor:
    for (iso3, meta, data) in data_list:
        cursor.execute("SELECT locationID FROM location WHERE iso3 = %s AND divisionname IS NULL AND city IS NULL", [iso3])
        locationid = list(cursor)[0][0]
        
        series_fields = "dataset, file, filehash, category, type, name, description"
        insert_series = "INSERT INTO SERIES (" + series_fields +") VALUES (%s, %s, %s, %s, %s, %s, %s)" 
        values = []
        for field in series_fields.split(","):
            values.append(meta[field.strip()])
        cursor.execute(insert_series, values)
        cursor.execute("SELECT LAST_INSERT_ID()")
        seriesid = list(cursor)[0][0]
        
        for date in data.index:
            cursor.execute("SELECT dateid FROM date WHERE year = %s AND month IS NULL AND day IS NULL", [str(date)])
            dateid = list(cursor)[0][0]
            observation_fields = "seriesid, locationid, dateid, value" 
            #full list of fields: `OBSERVATIONID`, `SERIESID`, `LOCATIONID`, `DATEID`, `NOTEID`, `DESCRIPTIONID`, `VALUE`, `UNITID`, `STATUSID`
            insert_observation = "INSERT INTO observation(" + observation_fields + ") VALUES (%s, %s, %s, %s)"
            values  = [seriesid, locationid, dateid, str(data.ix[date])]
            cursor.execute(insert_observation, values)
        cursor.execute("COMMIT")
```

Here is an example of how to make an SQL query to retrieve data in the series:

```python
df = ""
with smu.get_db_connection() as cursor:
    query = """
            SELECT D.year, L.iso3, O.value 
            FROM observation as O 
            INNER JOIN date as D    ON O.dateid = D.id
            INNER JOIN location as L   ON O.locationid = L.id
            INNER JOIN series as S 	ON O.seriesid = S.Id
            WHERE
                (L.iso3 = "TTO" OR
                L.iso3 = "BRB")
            AND S.dataset = 'UN DESA world population estimates'
            ORDER BY D.year, L.iso3
            """
    df = psql.frame_query(query, con=cursor.connection)

    
df
```
